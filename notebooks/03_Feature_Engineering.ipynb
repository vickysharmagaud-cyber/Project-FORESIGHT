```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv(
    "../data/processed/cleaned_retail_data.csv",
    parse_dates=["InvoiceDate"]
)

df.head()
```

```python
df["Date"] = df["InvoiceDate"].dt.date

df.head()
```

```python
daily_sales = (
    df.groupby(
        ["Date", "StockCode"]
    )
    .agg(
        Units_Sold=("Quantity", "sum"),
        Revenue=("Revenue", "sum")
    )
    .reset_index()
)

daily_sales.head()
```

```python
daily_sales["Date"] = pd.to_datetime(daily_sales["Date"])
```

```python
daily_sales = daily_sales.sort_values(
    ["StockCode", "Date"]
)
```

```python
daily_sales["Year"] = daily_sales["Date"].dt.year

daily_sales["Month"] = daily_sales["Date"].dt.month

daily_sales["Week"] = daily_sales["Date"].dt.isocalendar().week

daily_sales["Day"] = daily_sales["Date"].dt.day

daily_sales["DayOfWeek"] = daily_sales["Date"].dt.dayofweek
```

```python
daily_sales["Lag_1"] = (
    daily_sales.groupby("StockCode")["Units_Sold"]
    .shift(1)
)

daily_sales["Lag_7"] = (
    daily_sales.groupby("StockCode")["Units_Sold"]
    .shift(7)
)
```

```python
daily_sales["Rolling_7"] = (
    daily_sales.groupby("StockCode")["Units_Sold"]
    .transform(
        lambda x: x.rolling(7).mean()
    )
)
```

```python
daily_sales.dropna(inplace=True)
```

```python
daily_sales.to_csv(
    "../data/processed/daily_sales_features.csv",
    index=False
)

print("Feature dataset saved successfully!")
```
